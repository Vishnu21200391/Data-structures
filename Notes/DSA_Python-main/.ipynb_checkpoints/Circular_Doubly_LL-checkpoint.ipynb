```python
class Node:
    def __init__(self, prev = None, item = None, next = None):
        self.prev = prev
        self.item = item
        self.next = next


class CD_LL:
    def __init__(self, start=None):
        self.start = start

    def is_empty(self):
        return self.start is None

    def insert_at_beg(self,data):
        n = Node(None, data, self.start)
        if not self.is_empty(): 
            n.prev = self.start.prev
            self.start.prev.next = n
            self.start.prev = n
            self.start = n
            
        else:
            n.next = n
            n.prev = n
            self.start = n

    def insert_at_end(self,data):
        n = Node(None, data, None)
        if not self.is_empty(): 
            n.next = self.start
            n.prev = self.start.prev
            self.start.prev.next = n
            self.start.prev = n    
        else:
            n.next = n
            n.prev = n
            self.start = n

    def search(self,data):
        temp = self.start
        if not self.is_empty():
            while True:
                if temp.item == data:
                    return temp.item
                temp = temp.next
                if temp == self.start: 
                    break
        return None

    def insert_after(self, key, data):
        n = Node(None, data, None)
        temp = self.start
        if not self.is_empty():
            while True:
                if temp.item == key:
                    if temp.next == self.start:
                         n.next = self.start
                         n.prev = self.start.prev
                         self.start.prev.next = n
                         self.start.prev = n 
                    else:
                        n.prev = temp
                        n.next = temp.next
                        temp.next.prev = n
                        temp.next = n
                temp = temp.next
                        
                if temp == self.start:
                    break

    def delete_beg(self):
        if self.start.next == self.start:
            self.start = None
        else:
            self.start.next.prev = self.start.prev
            self.start.prev.next = self.start.next
            self.start = self.start.next

    def delete_last(self):
        if self.start.next == self.start:
            self.start = None
        else:
            self.start.prev.prev.next = self.start
            self.start.prev = self.start.prev.prev
            
    def delete_value(self, data):
        if self.is_empty():
            return
        temp = self.start

        while True:
            if temp.item == data:  
                if temp == self.start:
                    self.delete_beg()
                    temp = self.start
                    if self.is_empty():
                        break
                
                elif temp.next == self.start:
                    self.delete_last()
                    break
    
                elif temp.item == data:
                    temp.prev.next = temp.next
                    temp.next.prev = temp.prev
                    temp = temp.next
                    continue
                return
            temp = temp.next
    
            if temp == self.start:  
                break
  
    
    def display(self):
        if self.start is None:
            print("LL is empty")
        else:
            temp = self.start
            while True:
                print(temp.item, "->", end=" ")
                temp = temp.next
                if temp == self.start: 
                    break
            print("None")

cd = CD_LL()

cd.insert_at_beg(10)
cd.insert_at_beg(5)
cd.insert_at_beg(1)
cd.insert_at_end(15)
cd.insert_at_end(20)
cd.insert_after(cd.search(10), 19)
# cd.delete_beg()
# cd.delete_last()
# cd.delete_last()
# cd.delete_last()
# cd.delete_last()
# cd.delete_last()
cd.delete_value(cd.search(1))
cd.delete_value(cd.search(5))
cd.delete_value(cd.search(10))
cd.delete_value(cd.search(19))
cd.delete_value(cd.search(15))
cd.delete_value(cd.search(20))
cd.display()
```
