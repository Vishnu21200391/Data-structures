```python
import tkinter as tk
import random
import time
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

pause_flag = False
speed = 0.5
history = []  # Store previous states 

def draw_bars(canvas, arr, colors):
    canvas.delete("all")
    c_height = 300
    c_width = 600
    bar_width = c_width / len(arr)
    max_val = max(arr)
    for i, val in enumerate(arr):
        x0 = i * bar_width
        y0 = c_height - (val / max_val) * (c_height - 20)
        x1 = (i + 1) * bar_width
        y1 = c_height
        canvas.create_rectangle(x0, y0, x1, y1, fill=colors[i], outline="black")
        text_y = y0 - 15 if y0 > 25 else y0 + 10
        canvas.create_text((x0 + x1) / 2, text_y, text=str(val), fill="black", font=("Arial", 9, "bold"))
    canvas.update()

def wait_if_paused():
    while pause_flag:
        root.update()
        time.sleep(0.05)

def bubble_sort(canvas, arr):
    n = len(arr)
    start_time = time.time()
    for i in range(n):
        for j in range(n - i - 1):
            wait_if_paused()
            colors = ["red" if x == j or x == j + 1 else "skyblue" for x in range(n)]
            draw_bars(canvas, arr, colors)
            history.append(arr.copy())
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
            time.sleep(speed)
    draw_bars(canvas, arr, ["green" for _ in range(n)])
    return time.time() - start_time

def insertion_sort(canvas, arr):
    n = len(arr)
    start_time = time.time()
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            wait_if_paused()
            arr[j + 1] = arr[j]
            j -= 1
            colors = ["red" if x == j + 1 else "skyblue" for x in range(n)]
            draw_bars(canvas, arr, colors)
            history.append(arr.copy())
            time.sleep(speed)
        arr[j + 1] = key
    draw_bars(canvas, arr, ["green" for _ in range(n)])
    return time.time() - start_time

def selection_sort(canvas, arr):
    n = len(arr)
    start_time = time.time()
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            wait_if_paused()
            if arr[j] < arr[min_idx]:
                min_idx = j
            colors = ["red" if x == j or x == min_idx else "skyblue" for x in range(n)]
            draw_bars(canvas, arr, colors)
            history.append(arr.copy())
            time.sleep(speed)
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    draw_bars(canvas, arr, ["green" for _ in range(n)])
    return time.time() - start_time

def start_sorting(sort_func, canvas, arr, results):
    global pause_flag, history
    pause_flag = False
    history = []
    time_taken = sort_func(canvas, arr.copy())
    results[sort_func.__name__] = time_taken
    update_analyzer(results)

def update_analyzer(results):
    fig, ax = plt.subplots()
    ax.bar(results.keys(), results.values(), color=['blue', 'red', 'green'])
    ax.set_ylabel("Time (seconds)")
    ax.set_title("Sorting Algorithm Performance")
    canvas = FigureCanvasTkAgg(fig, master=root)
    canvas.get_tk_widget().pack()
    canvas.draw()

def pause_resume():
    global pause_flag
    pause_flag = not pause_flag
    pause_button.config(text="Resume" if pause_flag else "Pause")

def update_speed(val):
    global speed
    speed = float(val)

def main():
    global root, pause_button
    root = tk.Tk()
    root.title("Sorting Visualizer & Analyzer")
    canvas = tk.Canvas(root, width=600, height=300, bg="white")
    canvas.pack()
    arr = [random.randint(10, 100) for _ in range(20)]
    draw_bars(canvas, arr, ["skyblue" for _ in range(len(arr))])
    results = {}

    control_frame = tk.Frame(root)
    control_frame.pack()

    tk.Button(control_frame, text="Bubble Sort", command=lambda: start_sorting(bubble_sort, canvas, arr, results)).pack(side=tk.LEFT)
    tk.Button(control_frame, text="Insertion Sort", command=lambda: start_sorting(insertion_sort, canvas, arr, results)).pack(side=tk.LEFT)
    tk.Button(control_frame, text="Selection Sort", command=lambda: start_sorting(selection_sort, canvas, arr, results)).pack(side=tk.LEFT)
    tk.Button(control_frame, text="Shuffle", command=lambda: draw_bars(canvas, random.sample(arr, len(arr)), ["skyblue"] * len(arr))).pack(side=tk.LEFT)

    pause_button = tk.Button(control_frame, text="Pause", command=pause_resume)
    pause_button.pack(side=tk.LEFT)

    speed_scale = tk.Scale(control_frame, from_=0.01, to=2.0, resolution=0.01, label="Speed (s)", orient=tk.HORIZONTAL, command=update_speed)
    speed_scale.set(speed)
    speed_scale.pack(side=tk.LEFT)

    root.mainloop()

if __name__ == "__main__":
    main()
```
